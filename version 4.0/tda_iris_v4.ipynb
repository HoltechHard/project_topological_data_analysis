```python
# load the dataset
from sklearn.datasets import load_iris

# load iris dataset 
x, y = load_iris(return_X_y = True)
print("X: ", x[:10])
print("Y: ", y[:10])
```

```python
print("dim x = ", x.shape)
print("dim y = ", y.shape)
```

```python
# split the train-set and test-set
from sklearn.model_selection import train_test_split

# split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1, test_size = 0.3)
```

```python
# dimensions for each subset
print("dim x-train = ", x_train.shape)
print("dim y-train = ", y_train.shape)
print("dim x-test = ", x_test.shape)
print("dim y-test = ", y_test.shape)
```

```python
# calculation of persistence homology
from gtda.homology import VietorisRipsPersistence

persistence_homology = VietorisRipsPersistence(
    metric = "euclidean",
    homology_dimensions = [0, 1]
)
```

```python
# generate the diagrams of persistence
diagrams = persistence_homology.fit_transform([x, x_train, x_test])
```

```python
# check the vector of persistence features
print(diagrams)

# dimensions for vector of persistence features
print("dim persistence-diagrams = ", diagrams.shape)
```

```python
# split the persistence diagrams for each set of data
diagram_x = diagrams[0]
diagram_xtrain = diagrams[1]
diagram_xtest = diagrams[2]
```

```python
# plot the persistence diagram - for all dataset
from gtda.plotting import plot_diagram

plot_diagram(diagram_x)
```

```python
# plot the persistence diagram for x_train
plot_diagram(diagram_xtrain)
```

```python
# plot the persistence diagram for x_test
plot_diagram(diagram_xtest)
```

```python
# check the dimensionality for each persistence diagram
print("dim pers-diag-x = ", diagram_x.shape)
print("dim pers-diag-xtrain = ", diagram_xtrain.shape)
print("dim pers-diag-xtest = ", diagram_xtest.shape)
print("dim total pers-diagram = ", diagrams.shape)
```

```python
# calculate persistence vector of features 
from gtda.diagrams import PersistenceEntropy

persistence_entropy = PersistenceEntropy()
pe_x = persistence_entropy.fit_transform(diagrams)
```

```python
# plot the persistence entropy
from gtda.plotting import plot_point_cloud

plot_point_cloud(pe_x)
```

```python
print(pe_x)
```

```python
# compute the wasserstain distance
from gtda.diagrams import PairwiseDistance

wdis = PairwiseDistance(metric = "wasserstein", order = None)
```

```python
# wassertein distance between x and x_train
wdis_train = wdis.fit_transform([diagram_x, diagram_xtrain])
print(wdis_train)
print("dim wdis-train = ", wdis_train.shape)
```

```python
# wassertein distance between x and x_test
wdis_test = wdis.fit_transform([diagram_x, diagram_xtest])
print(wdis_test)
print("dim wdis-test = ", wdis_test.shape)
```

```python
# train the model
from sklearn.svm import SVC
classifier = SVC(gamma = "auto")
classifier.fit(wdis_train, y_train)
```
